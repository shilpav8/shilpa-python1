```python
# Constants
speed_thief = 100  # mph
speed_cop = 125  # mph
time_delay_minutes = 5  # minutes

# Convert time delay to hours
time_delay_hours = time_delay_minutes / 60

# Step 1: Distance covered by the thief during the delay
distance_thief = speed_thief * time_delay_hours

# Step 2: Relative speed between the cop and the thief
relative_speed = speed_cop - speed_thief

# Step 3: Time for the cop to catch the thief (in hours)
time_to_catch_hours = distance_thief / relative_speed

# Convert the time to minutes
time_to_catch_minutes = time_to_catch_hours * 60

# Output the results
print("Distance covered by the thief during the delay:", round(distance_thief,2), "miles")
print("Time for the cop to catch the thief:", round(time_to_catch_minutes,2), "minutes")
```

Probability

```python
#Conditional probability


# Import the sympy library for symbolic mathematics
import sympy as sp

# Define the total number of cards in the deck
total_cards = 52

# Define the number of Kings and Face cards
num_kings = 4
num_face_cards = 12

# Define the probabilities P(King ∩ Face Card) and P(Face Card)
P_King_and_Face = sp.Rational(num_kings, total_cards)  # P(King ∩ Face Card)
P_Face = sp.Rational(num_face_cards, total_cards)      # P(Face Card)

# Calculate the conditional probability P(King | Face Card)
P_King_given_Face = P_King_and_Face / P_Face

# Output the result
print(f"The probability of drawing a King given that it is a face card is: {P_King_given_Face}")
```

Joint probability

```python
# Import the sympy library for symbolic mathematics
import sympy as sp

# Define the probability of rolling a 6 on a single die
P_6 = sp.Rational(1, 6)

# Calculate the joint probability for both dice showing a 6
P_both_6 = P_6 * P_6

# Output the result
print(f"The probability of both dice showing a 6 is: {P_both_6}")
```

Bayes

```python
# Define the probabilities
P_A = 0.01  # Prior probability of having the disease
P_B_given_A = 0.95  # Probability of testing positive given that the person has the disease
P_B = 0.05  # Total probability of testing positive (including false positives)

# Apply Bayes' Theorem
P_A_given_B = (P_B_given_A * P_A) / P_B

# Output the result
print(f"The probability of having the disease given a positive test result is: {P_A_given_B:.4f}")
```

Binomial

```python
# Import necessary library
from scipy.stats import binom

# Define parameters
n = 10  # Number of trials (flips)
p = 0.5  # Probability of success (heads)
k = 6  # Number of successes (heads)

# Calculate the binomial probability of getting exactly 6 heads in 10 flips
probability = binom.pmf(k, n, p)

# Output the result
print(f"The probability of getting exactly 6 heads in 10 coin flips is: {probability:.2f}")
```
